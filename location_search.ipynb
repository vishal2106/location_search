```python
import pandas as pd
```

```python
df = pd.read_csv('/content/IN.txt', sep='\t', header=None)
df.head()
```

```python
header_dict = {
    0: 'geonameid',
    1: 'name',
    2: 'asciiname',
    3: 'alternatenames',
    4: 'latitude',
    5: 'longitude',
    6: 'feature class',
    7: 'feature code',
    8: 'country code',
    9: 'cc2',
    10: 'admin1 code',
    11: 'admin2 code',
    12: 'admin3 code',
    13: 'admin4 code',
    14: 'population',
    15: 'elevation',
    16: 'dem',
    17: 'timezone',
    18: 'modification date'
}
```

```python
df.rename(columns=header_dict, inplace=True)
df.head()
```

```python
null_counts = df.isnull().sum()

print("Number of null values in each column:")
print(null_counts)
print(len(df))
```

```python
# Dropping columns cc2, admin4 code, elevation due to large null values count
cols_to_drop = ['cc2', 'admin4 code', 'elevation']
df = df.drop(columns=cols_to_drop)
df.head()
```

```python
feature_class_values_counts = df['feature class'].value_counts()
feature_code_values_counts = df['feature code'].value_counts()
country_code_values_counts = df['country code'].value_counts()
admin1_code_values_counts = df['admin1 code'].value_counts()


print("Unique values of feature class:")
print(feature_class_values_counts)

print("Unique values of feature code:")
print(feature_code_values_counts)

print("Unique values of country code:")
print(country_code_values_counts)

print("Unique values of admin1 code:")
print(admin1_code_values_counts)
```

```python
# Replace feature class column data with their values

feature_class_mapping = {
    'A': 'state',
    'H': 'stream',
    'L': 'parks',
    'P': 'city',
    'R': 'road',
    'S': 'spot',
    'T': 'mountain',
    'U': 'undersea',
    'V': 'forest'
}

df['feature class'] = df['feature class'].replace(feature_class_mapping)
df.head()
```

```python
# Replace feature code column data with their values

feature_code_df = pd.read_csv('/content/feature_code.txt', sep='\t', header=None)
feature_code_header_dict = {
    0: 'feature code',
    1: 'feature code value',
    2: 'description'
}

state_code_df = pd.read_csv('/content/states.txt', sep='\t', header=None)
state_code_header_dict = {
    0: 'admin1 code',
    1: 'state'
}
```

```python
feature_code_df.rename(columns=feature_code_header_dict, inplace=True)
state_code_df.rename(columns=state_code_header_dict, inplace=True)
state_code_df['admin1 code'] = state_code_df['admin1 code'].astype(float)

state_code_df.head()
```

```python
feature_code_df = feature_code_df.drop(columns=['description'])

location_df = pd.merge(df, feature_code_df, left_on='feature code', right_on='feature code', how='left')
location_df.drop(columns=['feature code'], inplace=True)

location_df = pd.merge(df, state_code_df, left_on='admin1 code', right_on='admin1 code', how='left')
location_df.drop(columns=['admin1 code'], inplace=True)

location_df.head()
```

```python
# Drop columns which are not required
# Since country code has only one value (IN), we can drop these columns
# ['country code', 'admin2 code', 'admin3 code', 'population', 'elevation', 'dem', 'timezone', 'modification date']
location_df = location_df.drop(columns=['country code', 'admin2 code', 'admin3 code', 'population', 'dem', 'timezone', 'modification date'])
location_df.rename(columns={'feature code value': 'feature code'}, inplace=True)
location_df.head()
```

```python
state_value_counts = location_df['state'].value_counts()


print("Unique values of state:")
print(state_value_counts)
```

```python
location_df.to_json('location_data.json', orient='records')
```
